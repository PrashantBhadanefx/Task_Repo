### Model building

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
```

```python
# Using K- Means
```

```python
# Find the optimal number of clusters using Elbow Method
inertia = []
K_range = range(1, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(K_range, inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters ')
plt.ylabel('Inertia')
plt.show()
```

```python
# based on above graph 3 & 4 number is perfect number of clusters
```

```python
# evaluation
```

```python
silhouette_scores = []       # Score for each cluster
K = range(1, 11)
for k in K[1:]:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(df_scaled)
    score = silhouette_score(df_scaled, kmeans.labels_)
    silhouette_scores.append(score)

plt.plot(K[1:], silhouette_scores, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Method')
plt.show()
```

## Method 2

```python
from sklearn.decomposition import PCA

# Initialize PCA and fit to data
pca = PCA(n_components=5)  # Select the top 5 components
X_new = pca.fit_transform(df_scaled)

#  the amount of variance explained by each component
explained_variance = pca.explained_variance_ratio_

# Display the variance explained by each component
for i, var in enumerate(explained_variance):
    print(f'Component {i+1} explains {var:.2f} of the variance')
```

```python
# Plot explained variance
plt.figure(figsize=(10, 6))
plt.bar(range(1, len(explained_variance) + 1), explained_variance)
plt.title('Explained Variance by Each Principal Component')
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance')
plt.show()
```

```python
# Choose the number of components based on variance
pca = PCA(n_components=0.50)  #  50% of the variance
X_pca = pca.fit_transform(df_scaled)

print(f"Number of components chosen: {X_pca.shape[1]}")
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Perform K-means clustering on the PCA-transformed data
kmeans = KMeans(n_clusters=4, random_state=42)  # Assuming we want 4 clusters
kmeans.fit(X_pca)

# Get cluster labels
labels = kmeans.labels_

# Evaluate the clustering performance using Silhouette Score
sil_score = silhouette_score(X_pca, labels)
print(f"Silhouette Score for K-means clustering: {sil_score}")

# Add the cluster labels back to the original data

X_pca_df = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(X_pca.shape[1])])
X_pca_df['Cluster'] = labels


# Display the first few rows of the dataset with clusters
X_pca_df.head()
```

```python
# Visualizing the clusters in 2D (if we have 2 principal components)
plt.figure(figsize=(10, 6))
plt.scatter(X_pca_df['PC1'], X_pca_df['PC2'], c=X_pca_df['Cluster'], cmap='viridis', s=50, alpha=0.7)
plt.title('K-means Clusters in PCA-transformed Space (2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.show()
```
